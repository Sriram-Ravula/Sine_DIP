```python
from scipy.signal import chirp 
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2, 16384)
w = chirp(t, f0=500, f1=1, t1=2, method='linear')
print(w.shape)
plt.figure()
plt.plot(t, w)
plt.title("Linear Chirp, f(0)=6, f(10)=1")
plt.xlabel('t (sec)')
plt.show()

spectrum =np.fft.fft(w, norm='ortho')
spectrum = abs(spectrum[0:round(len(spectrum)/2)]) # Just first half of the spectrum, as the second is the negative copy

plt.figure()
plt.plot(spectrum, 'r')
plt.xlabel('Frequency (hz)')
plt.title('Original Waveform')
plt.xlim(0, 1200)
plt.show()
```

```python
import pandas as pd

x = pd.read_csv("/home/sravula/AirQualityUCI/AirQuality.csv")
one = x["O3-1"].dropna().values[0:1024]
two = x["O3-2"].dropna().values[0:1024]
three = x["NO2-1"].dropna().values[0:1024]
four = x["NO2-2"].dropna().values[0:1024]
five = x["CO-1"].dropna().values[0:1024]
six = x["CO-2"].dropna().values[0:1024]

def scale(signal):
    maxi = np.amax(signal)
    mini = np.amin(signal)
    ranges = maxi-mini
    y = 2*(signal - mini)/ranges - 1
    return y

def plotfft(signal, title):
    spectrum =np.fft.fft(signal, norm='ortho')
    spectrum = abs(spectrum[0:round(len(spectrum)/2)]) # Just first half of the spectrum, as the second is the negative copy

    plt.figure()
    plt.plot(spectrum, 'r')
    plt.xlabel('Frequency (hz)')
    plt.title(title)
    plt.xlim(0, 300)
    plt.show()

one = scale(one)
two = scale(two)
three = scale(three)
four = scale(four)
five = scale(five)
six = scale(six)

plotfft(one, "O3-1")
plotfft(two, "O3-2")
plotfft(three, "NO2-1")
plotfft(four, "NO2-2")
plotfft(five, "CO-1")
plotfft(six, "CO-2")
```
